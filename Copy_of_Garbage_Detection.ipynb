```python
!pip install zipfile
import os
import zipfile

# Replace 'path/to/predict_folder' with the actual path to your predict folder on Colab
folder_to_zip = '/content/runs/detect/predict'
output_filename = 'predict_folder.zip'

def zip_folder(folder_path, output_filename):
  """Zips a folder and saves it with the specified output filename.

  Args:
    folder_path: The path to the folder to zip.
    output_filename: The desired name of the output zip file.
  """
  with zipfile.ZipFile(output_filename, 'w') as zip_file:
    for root, _, files in os.walk(folder_path):
      for file in files:
        file_path = os.path.join(root, file)
        zip_file.write(file_path, os.path.relpath(file_path, folder_path))

zip_folder(folder_to_zip, output_filename)

print(f"Zipped 'predict' folder to '{output_filename}'.")

# Download the zipped folder using the files.download() function
from google.colab import files
files.download(output_filename)
```
